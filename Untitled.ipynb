```python
population_df = pd.read_csv("https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv")
```

```python
population_df = population_df[(population_df['state/region'] != 'USA') & (population_df.year == 2013) & (population_df.ages == 'total')
             ].rename(columns={'state/region': 'state'})[['state', 'population']]
```

```python
new_df = new_df.merge(population_df)
```

```python
new_df
```

```python
n_states = 9
top_states = list(population_df.sort_values(by="population", ascending=False).state[:n_states])
```

```python
top_states = list(new_df.groupby('state').total.sum().reset_index().sort_values(by='total', ascending=False).state[:n_states])
```

```python
`sorted_new_df = new_df.sort_values(by='date')
sorted_new_df[(sorted_new_df.state.isin(top_states)) & ((sorted_new_df.positive <= 0) | (sorted_new_df.total <= 0))]
```

```python
per_capita = pd.DataFrame({'state': new_df.state, 'date': new_df.date, 
                           'positive': new_df.positive / new_df.population,
                           'total': new_df.total / new_df.population,
                          })
```

```python
state_subset = per_capita[per_capita.state.isin(top_states)]
```

```python
hacked_total = state_subset.copy()
# hacked_total = hacked_total[(hacked_total.state == 'NY') & (hacked_total.positive > 0) & (hacked_total.total > 0)]
hacked_total = hacked_total[(hacked_total.positive > 0) & (hacked_total.total > 0)]
hacked_total[(hacked_total.positive <= 0)].positive = 1
hacked_total[(hacked_total.total <= 0)].total = 1
```

```python
hacked_total
```

```python
positive = alt.Chart().mark_line(strokeDash=[10,10]).encode(
    alt.X('date'),
    alt.Y('positive', scale=alt.Scale(type='log'), title="per capita"),
)
total = alt.Chart().mark_line().encode(
    alt.X('date'),
    alt.Y('total', scale=alt.Scale(type='log')),
)

title = ["new tests (solid line)", 
         "new positive results (dashed line)",
        "(source code: https://github.com/dchudz/something)",
        f"{n_states} most populous US states"]

alt.layer(positive, total
         ).facet(
    data=hacked_total,
    facet=alt.Facet('state', title=title),
                columns=4).configure_legend(symbolSize=120,
                                   labelLimit= 0,
                                   symbolType='circle',
                                   labelFontSize=14
                )

```
